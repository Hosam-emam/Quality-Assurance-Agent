```python
items = ['item 1', 'item 2', 'item 3', 'item 4']

import asyncio

# Iterate over the items of the list asynchronously
async def process_items(item):
    print(f'Processed: {item}')

    return f"Done {item}"
```

```python
async def process_items_concurrently(items):
    tasks = [process_items(item) for item in items]
    results = await asyncio.gather(*tasks)
    return results
```

```python
asyncio.run(process_items_concurrently(items))
```
